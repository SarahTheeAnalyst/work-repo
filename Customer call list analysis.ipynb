```python
import pandas as pd
```

```python
df = pd.read_excel("Downloads/Customer Call List.xlsx")
df
```

```python
df=df.drop_duplicates()
df
```

```python
df= df.drop(columns='Not_Useful_Column')
df
```

```python
#df["Last_Name"]= df['Last_Name'].str.lstrip('/')
#df["Last_Name"]= df["Last_Name"].str.lstrip('...')
#df['Last_Name']= df["Last_Name"].str.rstrip("_")
df['Last_Name'] = df['Last_Name'].str.strip("/._")
df
```

```python
df['Phone_Number']= df["Phone_Number"].astype(str).str.replace('[^a-zA-Z0-9]','',regex=True)
df
```

```python
#df['Phone_Number']= df['Phone_Number'].apply(lambda x: str(x))
#df
df['Phone_Number']= df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6]+ '-' + x[6:10])
df
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace('nan--',"")
df["Phone_Number"]= df['Phone_Number'].str.replace('Na--','')
df
```

```python
#df["Address"].str.split(',',n=2, expand=True)
df[['Sreet_Address','State','Zip_Code']] =df['Address'].str.split(',', expand=True)
df
```

```python
df=df.drop(columns="Address")
df
```

```python
#df['Paying Customer']= df['Paying Customer'].str.replace('Yes','Y')
#df['Paying Customer'] = df["Paying Customer"].str.replace('No','N')
#df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
#df['Do_Not_Contact'] =df['Do_Not_Contact'].str.replace('No','N')
#df["Paying Customer"]= df['Paying Customer'].replace({'Yes':'Y','No':'N'})
#df["Do_Not_Contact"]= df['Do_Not_Contact'].replace({'Yes':'Y','No':'N'})
df[["Paying Customer",'Do_Not_Contact']]= df[["Paying Customer",'Do_Not_Contact']].replace({'Yes':'Y','No':'N'})
df
```

```python
#df.replace({''})
df=df.fillna('')
df=df.replace('N/a',"")
df
```
